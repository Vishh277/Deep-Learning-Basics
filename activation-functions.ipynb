# **Activation Functions**

- Activation Function are the conditional statement of neural networks attached to each Neuron in ANN.
- AF provide useful information while suppressing Irrelevant Information to Data Points.
- It is also called as Transfer Function.

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
```

# 1) Linear Activation Function

- Activation is proportional to Input
- Used in Regression problems
- Used to learn Data Distribution

```python
def linear(x):
    return x
```

```python
x = np.linspace(-10,10)
plt.plot(x,linear(x))
```

# 2) Non Linear Activation Functions

## Sigmoid Function

- Used in Classification problems.
- Return Output between 0 - 1 irrespective of Input
- Mostly used in Outpuut Layer of Classification problems(Binary)

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
x = np.linspace(-10,10)
plt.plot(x,sigmoid(x))
plt.title('Sigmoid AF')
plt.show()
```

```python
after_sigmoid =  sigmoid(x)
after_sigmoid  # All  values between 0 and 1
```

```python
def sigmoid_derivation(x):  # Derivation of sigmoid
    return sigmoid(x) * (1-sigmoid(x))
```

```python
x = np.linspace(-10,10)
plt.plot(x,sigmoid_derivation(x))
plt.title('Sigmoid Derivation AF')
plt.show()
```

```python
plt.plot(x,sigmoid(x))
plt.plot(x,sigmoid_derivation(x))
plt.title('The Comparision')
plt.show()
```

## Tanh Activation

- Zero- centered output. 
- Output range -1 to 1
- Can be used in classification problmes

```python
def tanh(x):
    return np.tanh(x)
```

```python
x = np.linspace(-10,10)
plt.plot(x,tanh(x))
plt.title('TanH AF')
plt.show()
```

## ReLu AF

- Mostly used Hidden layer AF
- Neurons will be activated only if input is above 1
- Used to tackle Vanishing gradients

```python
def relu(x):
    x1 = []
    for i in x:
        if i<0:
            x1.append(0)
        else:
            x1.append(i)
    return x1
```

```python
x = np.linspace(-10,10)
plt.plot(x,relu(x))
plt.title('ReLu AF')
plt.show()
```

## Leaky ReLu AF

- Used to tackle Vanishing gradient on the negative side of ReLu AF
- Leaks 0.1 of value 

```python
def leaky_relu(x,alpha=0.1):
    dx = np.ones_like(x)
    dx[x<0] = alpha
    return dx
```

```python
x = np.linspace(-10,10)
plt.plot(leaky_relu(x,alpha=0.1))
plt.title('Leaky ReLu AF')
plt.show()
```

## Softmax Function

- Used for Multi-Class Classification Problems
- Mostly in Output Layer
- Similar to Sigmoid, but calculates relative probability

```python
def softmax(x):
    return np.exp(x)/np.sum(np.exp(x),axis=0)
```

```python
x = np.linspace(-10,10)

plt.plot(x,softmax(x))
plt.title('Softmax AF')
plt.show()
```
